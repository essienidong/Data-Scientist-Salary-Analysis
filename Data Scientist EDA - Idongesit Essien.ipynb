# Work Year Analysis

```python
#2.1 Remote Ratio by Work Year
#create pivot of Remote Ratio analysis and Work Year
Remote_ratio_pivot = pd.crosstab(df.remote_ratio,df.work_year,normalize='index')

#plot the data
Remote_ratio_pivot.plot.bar(stacked=True, figsize=(15,8))

plt.title('Remote Ratio Distribution by Work Year')
plt.show()
```

# Experience Level Analysis

```python
#Experience Level Analysis
#3.1. Employment Type by Experience Level
employment_type = pd.crosstab(df.employment_type,df.experience_level,normalize='index')

#plot the data
employment_type.plot.bar(stacked=True, figsize=(15,8))

plt.title('Employment Type Distribution by Experience Level')
plt.show()
```

```python
#3.2. Top 3 Job Title by Experience Level
df.groupby('job_title').experience_level.value_counts().nlargest(3)
```

```python
#3.2. Top 3 Job Title by Experience Level (Visualisation)
jbexp=df.groupby('job_title').experience_level.value_counts().nlargest(3)
#plot the data
jbexp.plot.bar(stacked=True, figsize=(15,8))
plt.title('Top 3 Job title by Experience Level')
plt.show()
```

```python
#3.3. Company Size by Experience Level
company_size = pd.crosstab(df.company_size,df.experience_level,normalize='index')

#plot the data
company_size.plot.bar(stacked=True, figsize=(15,8))

plt.title('Company Size Distribution by Experience Level')
plt.show()
```

```python
df.groupby('company_size').experience_level.value_counts()
```

# Salary Analysis

```python
5.1 # draw lineplot
sns.lineplot(x='work_year', y='salary', data=df)
# setting the title using Matplotlib
plt.title('Work_year and Salary')
plt.show()
```

```python
#5.2 draw lineplot Salary by Experience Level
sns.lineplot(x='experience_level', y='salary', data=df)
# setting the title using Matplotlib
plt.title('Salary Distribution by Experience Level')
plt.show()
```

```python
#5.3. Salary by Company Size
plt.figure(figsize=(10,10))
plt.scatter(df.company_size,df.salary)
plt.title('Salary Distribution by Company_size')
plt.show()
```

```python
#5.4 Highest salaries by job title
df.groupby('job_title').salary.max()
```

```python
#5.4 Highest salaries by job title
jobsal=df.groupby('job_title').salary.agg(max)
#plot the data
jobsal.plot.bar(stacked=True, figsize=(20,15))
plt.title('Highest Salaries by Job Title')
plt.show()
```

```python
#5.5. Average Salary by Company Location
df.groupby('company_location').salary.mean()
```

```python
#5.5. Average Salary by Company Location
salcoy=df.groupby('company_location').salary.mean()
#plot the data
salcoy.plot.bar(stacked=True, figsize=(20,15))
plt.title('Mean Salary by Company Location')
plt.show()
```
